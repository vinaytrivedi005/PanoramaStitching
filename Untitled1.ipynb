```python
from PIL import Image, ImageChops
im = Image.open('images/output.jpg')
print(im.size)
def trim(im):
    bg = Image.new(im.mode, im.size, im.getpixel((im.size[0]-1,0)))
    diff = ImageChops.difference(im, bg)
    diff = ImageChops.add(diff, diff, 2.0, -100)
    bbox = diff.getbbox()
    if bbox:
        return im.crop(bbox)
```

```python
trim(im).show()
```
